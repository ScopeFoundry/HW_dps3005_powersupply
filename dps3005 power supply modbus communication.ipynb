```python
import minimalmodbus
```

```python
instrument = minimalmodbus.Instrument('COM4', 1)
```

```python
instrument.serial.port
```

```python
instrument.serial.baudrate = 9600
instrument.serial.bytesize = 8
instrument.serial.timeout = 0.5
instrument.mode = minimalmodbus.MODE_RTU
```

```python
instrument.read_registers(0, 16)
```

```python
%timeit -r 1 -n 10 instrument.read_registers(0, 13)
```

```python
# Output Voltage (RO)
%timeit -r 1 -n 10 instrument.read_register(2, 2)
```

```python
# Output current (RO) Amps
instrument.read_register(3, number_of_decimals=3)
```

```python
# Power output (RO) in W
instrument.read_register(4, 2)
```

```python
# Volt input
instrument.read_register(5,2)
```

```python
# CV/CC status
instrument.read_register(8,0) 
# ->1 CC
# ->0 CV
```

```python
# Output enable read/writeable

instrument.read_register(9,0)
# 1 --> output enable
# 0 --> output disable
```

```python
# Model number (RO)
instrument.read_register(0xB,0)
```

```python
# Firmware version (RO)
instrument.read_register(0xC,0)
```

```python
#Voltage write
instrument.write_register(0, 4.25, 2)
```

```python
# Current Limit write in A
instrument.write_register(1, 0.8, 3)
```

```python
# Output enable
instrument.write_register(9,0,0)
instrument.read_register(9,0)
```

```python
# LED Brightness (0-5)
print(instrument.read_register(0xA,0))
instrument.write_register(0xA,5,0)
```

```python
# Lock input
print(instrument.read_register(6,0))
#instrument.write_register(6,0,0)
```
